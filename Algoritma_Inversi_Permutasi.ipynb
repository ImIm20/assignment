```python
input_sequence = input().split()
seq = [int(x) for x in input_sequence]
seq
```

```python
input_invers = input().split()
inversion = [int(x) for x in input_invers]
inversion 
```

```python
res = [max(seq)]
if len(seq) != len(inversion):
  print("the number of input must have the same length")
else:
  for i in range(len(seq)-2, -1, -1):
    res.insert(inversion[i], seq[i])
res
```
